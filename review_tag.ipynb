```python
stop_words = []
with open('stoplist.txt','r',encoding='utf-8') as st:
    for i in st.readlines():
        i = i.rstrip()
        stop_words.append(i)
```

```python
import jieba.posseg as pseg
reviews_attri = []
for i in reviews:
    words = pseg.cut(i)
#     texts = [flag for word,flag in words]
#     reviews_attri.append(texts)
    try:
        for word,flag in words:
            if word not in stop_words:
                reviews_line = []
                reviews_line.append(word)
                reviews_line.append(flag)
                reviews_attri.append(reviews_line)
    except AttributeError:
        pass
```

```python
length = len(reviews_attri)
reviews_tag = ''
for i in range(0,length):
    val1 = reviews_attri[i][1]
    key1 = reviews_attri[i][0]
    try:
        if val1 == 'v':
            val2 = reviews_attri[i+1][1]
            key2 = reviews_attri[i+1][0]
            if val2 == 'a' or 'n' or 'c':
#                 print (key1,key2)
                opionion_phase = str(key1 + key2)
                reviews_tag = reviews_tag + '，' + opionion_phase
        elif val1 == 'n' or 'ns':
            val2 = reviews_attri[i+1][1]
            val3 = reviews_attri[i-1][1]
            key2 = reviews_attri[i+1][0]
            if val2 == 'a' and val3 != 'p' and val3 != 'c':
#                 print (key1,key2)
                opionion_phase = str(key1 + key2)
                reviews_tag = reviews_tag + '，' + opionion_phase
        elif val1 == 'p' or 'c':
            val2 = reviews_attri[i+1][1]
            key2 = reviews_attri[i+1][0]
            if val2 == 'eng' or 'n':
                key3 = reviews_attri[i+2][0]
#                 print (key1,key2,key3)
                opionion_phase = str(key1 + key2 + key3)
                reviews_tag = reviews_tag + '，' + opionion_phase
    except IndexError:
        pass
```
